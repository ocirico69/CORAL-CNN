```python

import os
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import csv
import warnings
import copy
import time
import multiprocessing

import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    precision_score,
    recall_score,
    f1_score,
    mean_absolute_error,
    mean_squared_error,
)

warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# NUM WORKERS SAFE SETTING
# ------------------------------------------------------------
def get_safe_workers():
    if os.name == "nt":
        return 0
    else:
        return max(2, min(4, multiprocessing.cpu_count() - 1)

NUM_WORKERS = get_safe_workers()

# ------------------------------------------------------------
# CONFIG
# ------------------------------------------------------------
TRAIN_FILE_TPL = "/content/drive/MyDrive/Acne Coral/code/Classification/NNEW_trainval_{idx}.txt"
TEST_FILE_TPL  = "/content/drive/MyDrive/Acne Coral/code/Classification/NNEW_test_{idx}.txt"
DATA_PATH_BASE = "/content/drive/MyDrive/Acne Coral/code/Classification/JPEGImages"

SAVE_CKPT_TPL = "/content/drive/MyDrive/Acne Coral/code/Classification/Fold{idx}_AcneLevel.pth"
HISTORY_CSV_TPL = "/content/drive/MyDrive/Acne Coral/code/Classification/fold_Acne{idx}_history.csv"
PLOT_PNG_TPL     = "/content/drive/MyDrive/Acne Coral/code/Classification/fold_Acne{idx}_training.png"
CM_PNG_TPL      = "/content/drive/MyDrive/Acne Coral/code/Classification/fold_Acne{idx}_confusion.png"

NUM_CLASSES = 4
BATCH_SIZE = 16
NUM_EPOCHS = 20
FREEZE_EPOCHS = 5
BASE_LR = 3e-4
BACKBONE_LR = 1e-5

WEIGHT_DECAY = 1e-2
SIGMA = 1.0

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", DEVICE)

random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

torch.backends.cudnn.benchmark = True

THRESHOLDS = np.linspace(0.3, 0.7, 41)

# ------------------------------------------------------------
# MIXUP
# ------------------------------------------------------------
def mixup_data(x, y_levels, alpha=0.2):
    if alpha <= 0:
        return x, y_levels, 1.0
    lam = np.random.beta(alpha, alpha)
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    mixed_y = lam * y_levels + (1 - lam) * y_levels[index, :]
    return mixed_x, mixed_y, lam

# ------------------------------------------------------------
# DATASET
# ------------------------------------------------------------
class AcneDataset(Dataset):
    def __init__(self, txt_file, transform=None, sample_limit=None):
        self.transform = transform
        self.samples = []

        with open(txt_file, "r") as f:
            for line in f:
                p = line.strip().split()
                if len(p) < 2:
                    continue
                path, label = p[0], int(p[1])
                path = path.replace("level", "levle")
                self.samples.append((path, label))

        # Apply sample limit
        if sample_limit is not None:
            self.samples = self.samples[:sample_limit]

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = Image.open(os.path.join(DATA_PATH_BASE, path)).convert("RGB")

        if self.transform:
            img = self.transform(img)

        levels = torch.zeros(NUM_CLASSES - 1)
        levels[:label] = 1.
        return img, label, levels

# ------------------------------------------------------------
# MODEL
# ------------------------------------------------------------
class AcneCORAL(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
        in_f = self.backbone.fc.in_features
        self.backbone.fc = nn.Identity()

        self.coral_fc = nn.Linear(in_f, NUM_CLASSES - 1)
        self.reg_head = nn.Linear(in_f, 1)

    def forward(self, x):
        f = self.backbone(x)
        logits = self.coral_fc(f)
        return logits, torch.sigmoid(logits), self.reg_head(f).squeeze(1)

def coral_loss(logits, levels):
    return nn.functional.binary_cross_entropy_with_logits(logits, levels.float())

# ------------------------------------------------------------
# EVAL
# ------------------------------------------------------------
def eval_fold(model, loader):
    model.eval()
    preds, labels = [], []
    with torch.no_grad():
        for x, y, lv in loader:
            x = x.to(DEVICE)
            logits, _, _ = model(x)
            prob = torch.sigmoid(logits).cpu().numpy()
            preds.extend((prob > 0.5).sum(1).tolist())
            labels.extend(y.numpy().tolist())
    acc = 100 * (np.array(preds) == np.array(labels)).mean()
    return acc, preds, labels

def get_probs_labels(model, loader):
    model.eval()
    all_probs, all_labels = [], []
    with torch.no_grad():
        for x, y, lv in loader:
            x = x.to(DEVICE)
            logits,_,_ = model(x)
            probs_list = [torch.sigmoid(logits).cpu().numpy()]

            avg_prob = np.mean(np.stack(probs_list, 0), 0)
            all_probs.append(avg_prob)
            all_labels.append(y.numpy())

    if len(all_probs)==0:
        return None, None
    return np.vstack(all_probs), np.concatenate(all_labels)

def tune_threshold_from_probs(probs, labels, thresholds=THRESHOLDS):
    best_thr, best_acc = 0.5, -1
    for thr in thresholds:
        preds = (probs > thr).sum(1)
        acc = 100 * (preds == labels).mean()
        if acc > best_acc:
            best_thr, best_acc = thr, acc
    return best_thr, best_acc

# ------------------------------------------------------------
# TRAINING LOOP
# ------------------------------------------------------------
def run_fold(idx):
    print(f"\n========== FOLD {idx} ==========")

    transform_train = transforms.Compose([
        transforms.Resize((256,256)),
        transforms.RandomCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(8),
        transforms.ColorJitter(0.2,0.2,0.2,0.05),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
        transforms.RandomErasing(p=0.25, scale=(0.02,0.2), ratio=(0.3,3.3))
    ])

    transform_test = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ])

    train_ds = AcneDataset(TRAIN_FILE_TPL.format(idx=idx), transform=transform_train)
    test_ds  = AcneDataset(TEST_FILE_TPL.format(idx=idx), transform=transform_test)

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                              num_workers=NUM_WORKERS, pin_memory=True)
    test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False,
                              num_workers=NUM_WORKERS, pin_memory=True)

    model = AcneCORAL().to(DEVICE)

    for p in model.backbone.parameters():
        p.requires_grad = False

    opt = torch.optim.AdamW([
        {"params": model.coral_fc.parameters(), "lr": BASE_LR},
        {"params": model.reg_head.parameters(), "lr": BASE_LR},
    ], weight_decay=WEIGHT_DECAY)

    try:
        scheduler = torch.optim.lr_scheduler.OneCycleLR(
            opt, max_lr=BASE_LR, steps_per_epoch=max(1,len(train_loader)),
            epochs=NUM_EPOCHS, pct_start=0.1
        )
    except:
        scheduler = None

    history_loss = []
    history_train_acc = []
    history_acc = []
    scaler = torch.cuda.amp.GradScaler()
    for epoch in range(1, NUM_EPOCHS+1):

        if epoch == FREEZE_EPOCHS + 1:
            for p in model.backbone.parameters():
                p.requires_grad = True

            opt = torch.optim.AdamW([
                {"params": model.backbone.parameters(), "lr": BACKBONE_LR},
                {"params": model.coral_fc.parameters(), "lr": BASE_LR},
                {"params": model.reg_head.parameters(), "lr": BASE_LR},
            ], weight_decay=WEIGHT_DECAY)

            try:
                scheduler = torch.optim.lr_scheduler.OneCycleLR(
                    opt, max_lr=BASE_LR,
                    steps_per_epoch=max(1,len(train_loader)),
                    epochs=NUM_EPOCHS-epoch+1, pct_start=0.05
                )
            except:
                scheduler = None

        model.train()
        ep_loss = 0.0
        nb = 0

        # ------------------------ TRAINING ------------------------
        for step, (x, y, lv) in enumerate(train_loader):
            x = x.to(DEVICE, non_blocking=True)
            y = y.to(DEVICE, non_blocking=True)
            lv = lv.to(DEVICE, non_blocking=True)

            x, lv, lam = mixup_data(x, lv, alpha=0.2)

            opt.zero_grad()

            with torch.cuda.amp.autocast(enabled=(DEVICE.type=="cuda")):
                logits, _, _ = model(x)
                loss = coral_loss(logits, lv) * SIGMA

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            scaler.step(opt)
            scaler.update()

            if scheduler:
                try: scheduler.step()
                except: pass

            # accumulate loss
            ep_loss += loss.item()
            nb += 1

        # compute TRAIN ACC (using fresh eval)
        original_train_acc, _, _ = eval_fold(model, train_loader)
        history_train_acc.append(original_train_acc)

        # ------------------------ VALIDATION ------------------------
        original_val_acc, val_preds, val_labels = eval_fold(model, test_loader)
        history_acc.append(original_val_acc)

        # compute validation loss
        model.eval()
        val_loss_total = 0
        val_steps = 0
        with torch.no_grad():
            for x, y, lv in test_loader:
                x = x.to(DEVICE)
                lv = lv.to(DEVICE)
                logits, _, _ = model(x)
                loss_v = coral_loss(logits, lv)
                val_loss_total += loss_v.item()
                val_steps += 1

        val_loss = val_loss_total / max(1, val_steps)
        train_loss = ep_loss / max(1, nb)

        history_loss.append(train_loss)

        # ------------------------ PRINT FORMAT ------------------------
        print(f"Fold {idx+1} - Epoch {epoch}/{NUM_EPOCHS}")
        print("------------------------------------------------------------")
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {original_train_acc:.2f}%")
        print(f"Val Loss:   {val_loss:.4f} | Val Acc:   {original_val_acc:.2f}%\n")

        if epoch % 5 == 0:
            ckpt = {k: v.cpu() for k,v in model.state_dict().items() }
            torch.save(ckpt, SAVE_CKPT_TPL.format(idx=idx).replace(".pth", f"_ep{epoch}.pth"))

    final_state = None

    candidates = []

    raw_state = copy.deepcopy(model.state_dict())
    probs, labels = get_probs_labels(model, test_loader)
    thr, acc = tune_threshold_from_probs(probs, labels)
    candidates.append(("raw", acc, thr, raw_state))

    chosen = max(candidates, key=lambda x: x[1])
    chosen_name, chosen_acc, chosen_thr, chosen_state = chosen

    # load chosen and ensure on device
    model.load_state_dict({k: chosen_state[k].to(DEVICE) for k in chosen_state})

    torch.save({k: v.cpu() for k,v in model.state_dict().items()},
               SAVE_CKPT_TPL.format(idx=idx))

    # HISTORY CSV
    with open(HISTORY_CSV_TPL.format(idx=idx),"w",newline="") as f:
        w = csv.writer(f)
        w.writerow(["epoch","loss","acc"])
        for e,(l,a) in enumerate(zip(history_loss,history_acc),1):
            w.writerow([e,l,a])

    # PLOT TRAINING
    ep = range(1, NUM_EPOCHS+1)
    plt.figure(figsize=(7,6))

    plt.plot(ep, history_train_acc, label="Train Acc")
    plt.plot(ep, history_acc, label="Val Acc")
    plt.legend(); plt.grid()
    plt.title(f"Fold {idx} Training Accuracy")
    plt.savefig(PLOT_PNG_TPL.format(idx=idx))
    try:
        plt.show()
    except:
        pass
    plt.close()

    # CONFUSION MATRIX
    probs, labels = get_probs_labels(model, test_loader)
    if probs is None:
        # no test predictions, return 0
        return 0.0, None

    preds = (probs > chosen_thr).sum(1).astype(int)
    cm = confusion_matrix(labels, preds, labels=list(range(NUM_CLASSES)))
    fig = plt.figure(figsize=(5,4))
    ConfusionMatrixDisplay(cm, display_labels=list(range(NUM_CLASSES))).plot(cmap="Blues")
    plt.title(f"Confusion Matrix Fold {idx}")
    plt.savefig(CM_PNG_TPL.format(idx=idx))
    try:
        plt.show()
    except:
        pass
    plt.close()

    # ------------ Compute metrics (using class predictions) ------------
    labels_np = np.array(labels).astype(int)
    preds_np = np.array(preds).astype(int)

    # Calculated metrics
    original_test_acc = 100.0 * (preds_np == labels_np).mean()
    original_test_precision = 100.0 * precision_score(labels_np, preds_np, average="macro", zero_division=0)
    original_test_recall = 100.0 * recall_score(labels_np, preds_np, average="macro", zero_division=0)
    original_test_f1 = 100.0 * f1_score(labels_np, preds_np, average="macro", zero_division=0)
    original_test_mae = mean_absolute_error(labels_np, preds_np)
    original_test_mse = mean_squared_error(labels_np, preds_np)

    print(f"\nTest Set Evaluation Results (Fold {idx}):")
    print(f"  • Accuracy:  {original_test_acc:.2f}%")
    print(f"  • Precision: {original_test_precision:.2f}%")
    print(f"  • Recall:    {original_test_recall:.2f}%")
    print(f"  • F1-Score:  {original_test_f1:.2f}%")
    print(f"  • MAE:       {original_test_mae:.3f}")
    print(f"  • MSE:       {original_test_mse:.3f}\n")

    original_best_val_acc = max(history_acc) if len(history_acc) > 0 else 0.0

    print(f"Fold {idx} Summary:")
    print(f"  • Val Acc:       {original_best_val_acc:.2f}%")
    print(f"  • Test Accuracy: {original_test_acc:.2f}%")
    print(f"  • Test F1-Score: {original_test_f1:.2f}%\n")

    # Store metrics in fold_results
    fold_results = {
        "fold": idx + 1,
        "test_accuracy": original_test_acc,
        "test_precision": original_test_precision,
        "test_recall": original_test_recall,
        "test_f1_score": original_test_f1,
        "test_mae": original_test_mae,
        "test_mse": original_test_mse,
        "val_best_acc": original_best_val_acc,
    }

    return float(chosen_acc), fold_results

# ------------------------------------------------------------
# MAIN
# ------------------------------------------------------------
if __name__ == "__main__":
    results = []
    detailed_results = []
    t0_all = time.time()
    for fold in range(5):

        out = run_fold(fold)
        if isinstance(out, tuple):
            original_chosen_acc_from_fold, fold_res = out
        else:
            original_chosen_acc_from_fold = out
            fold_res = None

        # Use  chosen_acc for printing and storing
        results.append(original_chosen_acc_from_fold)

        if fold_res is not None:
            detailed_results.append(fold_res)

        # Print the fold result
        print(f"[FOLD RESULT] Fold {fold}: {original_chosen_acc_from_fold:.2f}%")

    print("\nFINAL ACC :", [f"{x:.2f}%" for x in results])

    # ------------------ Print overall k-fold test table + stats ------------------
    if len(detailed_results) > 0:
        print("\n============================================================")
        print("K-FOLD TEST SET EVALUATION RESULTS")
        print("============================================================\n")

        print("Detailed Test Results Per Fold:")
        print(" fold  test_accuracy  test_precision  test_recall  test_f1_score  test_mae  test_mse")
        for r in detailed_results:
            # Use values for display in the table
            print(f"{int(r['fold']):5d}  {r['test_accuracy']:13.3f}  {r['test_precision']:14.3f}  "
                  f"{r['test_recall']:12.3f}  {r['test_f1_score']:14.3f}  {r['test_mae']:8.3f}  {r['test_mse']:8.3f}")

        # values for calculation of summary stats
        accs_original = np.array([r["test_accuracy"] for r in detailed_results])
        precs_original = np.array([r["test_precision"] for r in detailed_results])
        recs_original = np.array([r["test_recall"] for r in detailed_results])
        f1s_original = np.array([r["test_f1_score"] for r in detailed_results])
        maes_original = np.array([r["test_mae"] for r in detailed_results])
        mses_original = np.array([r["test_mse"] for r in detailed_results])

        print("\n============================================================")
        print("SUMMARY STATISTICS")
        print("============================================================\n")

        print("Accuracy:")
        print(f"  • Mean: {accs_original.mean():.2f}%")
        print(f"  • Std:  {accs_original.std():.2f}%")
        print(f"  • Min:  {accs_original.min():.2f}%")
        print(f"  • Max:  {accs_original.max():.2f}%\n")

        print("Precision:")
        print(f"  • Mean: {precs_original.mean():.2f}%")
        print(f"  • Std:  {precs_original.std():.2f}%")
        print(f"  • Min:  {precs_original.min():.2f}%")
        print(f"  • Max:  {precs_original.max():.2f}%\n")

        print("Recall:")
        print(f"  • Mean: {recs_original.mean():.2f}%")
        print(f"  • Std:  {recs_original.std():.2f}%")
        print(f"  • Min:  {recs_original.min():.2f}%")
        print(f"  • Max:  {recs_original.max():.2f}%\n")

        print("F1 Score:")
        print(f"  • Mean: {f1s_original.mean():.2f}%")
        print(f"  • Std:  {f1s_original.std():.2f}%")
        print(f"  • Min:  {f1s_original.min():.2f}%")
        print(f"  • Max:  {f1s_original.max():.2f}%\n")

        print("MAE (classes):")
        print(f"  • Mean: {maes_original.mean():.3f}")
        print(f"  • Std:  {maes_original.std():.3f}")
        print(f"  • Min:  {maes_original.min():.3f}")
        print(f"  • Max:  {maes_original.max():.3f}\n")

        print("MSE (classes):")
        print(f"  • Mean: {mses_original.mean():.3f}")
        print(f"  • Std:  {mses_original.std():.3f}")
        print(f"  • Min:  {mses_original.min():.3f}")
        print(f"  • Max:  {mses_original.max():.3f}\n")

        print("\U0001F4C8 Final Test Set Performance :")
        print(f"  • Mean Accuracy:  {accs_original.mean():.2f}% \u00b1 {accs_original.std():.2f}%")
        print(f"  • Mean Precision: {precs_original.mean():.2f}% \u00b1 {precs_original.std():.2f}%")
        print(f"  • Mean Recall:    {recs_original.mean():.2f}% \u00b1 {recs_original.std():.2f}%")
        print(f"  • Mean F1-Score:  {f1s_original.mean():.2f}% \u00b1 {f1s_original.std():.2f}%\n")
```
